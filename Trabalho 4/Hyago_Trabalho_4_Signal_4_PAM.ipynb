> 1 - Um código para criação de um sinal 4-pam

```python
import numpy as np
import matplotlib.pyplot as plt

# Número de símbolos
N = 10000

# Geração de símbolos aleatórios entre 0 e 3
symbols = np.random.randint(0, 4, N)

"""
***********   4 - PAM - 4 níveis de amplitude    ***********

    -------------------------------------------------
                nível / símbolo -> amplitude

                  0   00           ->   -3
                  1   01           ->   -1
                  2   10           ->    1
                  3   11           ->    3

"""
amplitudes = {0: -100, 1: -50, 2: 50, 3: 100}
signaltrem = [amplitudes[symbol] for symbol in symbols]
# repetindo 5 amostras seguidas por o símbolo no mínimo
signal = np.repeat(signaltrem,5)
signaltrem = np.repeat(signaltrem,5)

# Plot do sinal mostrando 500 amostras
plt.figure(figsize=(10,4))
plt.stem(signaltrem[:250])
plt.title('Sinal 4-PAM - trem de impulsos')
plt.xlabel('Símbolo')
plt.ylabel('Amplitude')
plt.grid(True)
plt.show()


# Plot do sinal mostrando 500 amostras
plt.figure(figsize=(10,4))
plt.plot(signal[:250])
plt.title('Sinal 4-PAM')
plt.xlabel('Símbolo')
plt.ylabel('Amplitude')
plt.grid(True)
plt.show()
```

> 2 - Um código para demodução do sinal 4 -pam, considerar que o sinal passou por um canal ruidoso.

  - Depende da amplitude e se utiliza filtro no caso, porque veja abaixo caso a faixa da amplitude seja de -3 -1 1 3 o ruído altera o pulso e demodular se torna um desafio grande

  - Agora se a amplitude é de -100 -50 50 100 no caos o ruído não consegue alterar, mesmo que a amplitude do ruído seja alta, problema é quando o ruído é quase o valor de amplitude do pulso

  - o limiar também de que verifica para alterar os níveis é algo que altera também muito sobre o valor e como será demodulado, caso o limiar esteja muito perto dos máximos ou perto de algum nível isso irá alterar muito o sinal demodulado

```python
import numpy as np
import matplotlib.pyplot as plt

import numpy as np
import matplotlib.pyplot as plt

# Número de símbolos
N = 10000

# Geração de símbolos aleatórios entre 0 e 3
symbols = np.random.randint(0, 4, N)

"""
***********   4 - PAM - 4 níveis de amplitude    ***********

    -------------------------------------------------
                nível / símbolo -> amplitude

                  0   00           ->   -3
                  1   01           ->   -1
                  2   10           ->    1
                  3   11           ->    3

"""
amplitudes = {0: -3, 1: -1, 2: 1, 3: 3}
signaltrem = [amplitudes[symbol] for symbol in symbols]
# repetindo 5 amostras seguidas por o símbolo no mínimo
signal = np.repeat(signaltrem,10)
signaltrem = np.repeat(signaltrem,10)



# Ruído no canal
noise = np.random.normal(0, 1, len(signal))
# Sinal após passar pelo canal
received_signal = np.array(signal) + noise

# Limiares para a decisão
thresholds = [-2, 0, 2]

# Demodulação do sinal
demodulated_symbols = []
for value in received_signal:
    if value < thresholds[0]:
        demodulated_symbols.append(0)
    elif value < thresholds[1]:
        demodulated_symbols.append(1)
    elif value < thresholds[2]:
        demodulated_symbols.append(2)
    else:
        demodulated_symbols.append(3)

# Mapeamento dos símbolos demodulados de volta para as amplitudes
amplitudes = {0: -3, 1: -1, 2: 1, 3: 3}
demodulated_signal = [amplitudes[symbol] for symbol in demodulated_symbols]

erro = demodulated_signal - signal

# Plot do sinal mostrando 100 amostras
plt.figure(figsize=(10,4))
plt.stem(signaltrem[:250])
plt.title('Sinal 4-PAM - trem de impulsos')
plt.xlabel('Símbolo')
plt.ylabel('Amplitude')
plt.grid(True)
plt.show()


# Plot do sinal mostrando 500 amostras
plt.figure(figsize=(10,4))
plt.plot(signal[:250])
plt.title('Sinal 4-PAM')
plt.xlabel('Símbolo')
plt.ylabel('Amplitude')
plt.grid(True)
plt.show()

# Plot do ruído
plt.figure(figsize=(10,4))
plt.plot(noise[:250])
plt.title('Ruído')
plt.xlabel('Símbolo')
plt.ylabel('Amplitude')
plt.grid(True)
plt.show()

# Plot do sinal com ruído
plt.figure(figsize=(10,4))
plt.plot(received_signal[:250])
plt.title('Sinal 4-PAM com Ruído')
plt.xlabel('Símbolo')
plt.ylabel('Amplitude')
plt.grid(True)
plt.show()



# Plot do sinal demodulado
plt.figure(figsize=(10,4))
plt.plot(demodulated_signal[:250])
plt.title('Sinal Demodulado')
plt.xlabel('Símbolo')
plt.ylabel('Amplitude')
plt.grid(True)
plt.show()


# Plot do sinal demodulado
plt.figure(figsize=(10,4))
plt.plot(erro[:250])
plt.title('Erro')
plt.xlabel('Símbolo')
plt.ylabel('Amplitude')
plt.grid(True)
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

import numpy as np
import matplotlib.pyplot as plt

# Número de símbolos
N = 10000

# Geração de símbolos aleatórios entre 0 e 3
symbols = np.random.randint(0, 4, N)

amplitudes = {0: -100, 1: -50, 2: 50, 3: 100}
signaltrem = [amplitudes[symbol] for symbol in symbols]
# repetindo 5 amostras seguidas por o símbolo no mínimo
signal = np.repeat(signaltrem,5)
signaltrem = np.repeat(signaltrem,5)


# Ruído no canal
noise = 10*np.random.normal(0, 1, len(signal))
# Sinal após passar pelo canal
received_signal = np.array(signal) + noise

# Limiares para a decisão
thresholds = [-75, 0, 75]

# Demodulação do sinal
demodulated_symbols = []
for value in received_signal:
    if value < thresholds[0]:
        demodulated_symbols.append(0)
    elif value < thresholds[1]:
        demodulated_symbols.append(1)
    elif value < thresholds[2]:
        demodulated_symbols.append(2)
    else:
        demodulated_symbols.append(3)

# Mapeamento dos símbolos demodulados de volta para as amplitudes
amplitudes = {0: -100, 1: -50, 2: 50, 3: 100}
demodulated_signal = [amplitudes[symbol] for symbol in demodulated_symbols]


error = demodulated_signal - signal

# Plot do sinal mostrando 100 amostras
plt.figure(figsize=(10,4))
plt.stem(signaltrem[:250])
plt.title('Sinal 4-PAM - trem de impulsos')
plt.xlabel('Símbolo')
plt.ylabel('Amplitude')
plt.grid(True)
plt.show()


# Plot do sinal mostrando 500 amostras
plt.figure(figsize=(10,4))
plt.plot(signal[:250])
plt.title('Sinal 4-PAM')
plt.xlabel('Símbolo')
plt.ylabel('Amplitude')
plt.grid(True)
plt.show()

# Plot do ruído
plt.figure(figsize=(10,4))
plt.plot(noise[:250])
plt.title('Ruído')
plt.xlabel('Símbolo')
plt.ylabel('Amplitude')
plt.grid(True)
plt.show()

# Plot do sinal com ruído
plt.figure(figsize=(10,4))
plt.plot(received_signal[:250])
plt.title('Sinal 4-PAM com Ruído')
plt.xlabel('Símbolo')
plt.ylabel('Amplitude')
plt.grid(True)
plt.show()



# Plot do sinal demodulado
plt.figure(figsize=(10,4))
plt.plot(demodulated_signal[:250])
plt.title('Sinal Demodulado')
plt.xlabel('Símbolo')
plt.ylabel('Amplitude')
plt.grid(True)
plt.show()


# Plot do sinal demodulado
plt.figure(figsize=(10,4))
plt.plot(error[:250])
plt.title('Erro')
plt.xlabel('Símbolo')
plt.ylabel('Amplitude')
plt.ylim(-100,100)
plt.grid(True)
plt.show()

```
