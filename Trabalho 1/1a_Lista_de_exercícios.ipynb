de 6 falhas a cada duas semanas, calcule a probabilidade de que haverá ao menos 2 falhas durante uma
semana específica. Traçar o histograma da variável analisada.

```python
import math

lambda_ = 3

def poisson_probability(lambda_, k):
    return (lambda_**k * math.exp(lambda_*-1)) / math.factorial(k)

probabilidade = 1 - sum(poisson_probability(lambda_, k) for k in range(0,2))

print(f"A probabilidade de receber menos de oito chamadas em uma hora é {probabilidade} %")
```

 A distribuição discreta geométrica conta o número de tentativas até o primeiro sucesso. A pmf é dada
por f(x)=p(1-p)x-1
, onde p representa a probabilidade de sucesso e x o numero de tentativas. Fazer um
algoritmo para a geração das variáveis aleatórias geométricas. Com o algoritmo proposto calcular:
Uma urna tem 30 bolas brancas e 20 bolas pretas. Qual a probabilidade de que a 6ª bola
retirada com reposição seja a primeira bola preta?

```python
psucces = 20/50
xtry = 6

result = psucces*(1-psucces)**(xtry-1)

print(f'Resultado é : {result}')
```

```python
import numpy as np

p = 20/50
N = 100000
x = np.random.uniform(0, 1, N)
value = 6
count = 0

for ix in x:
    i = 1
    pr = p

    while ix >= pr:
        pr += p*(1 - p)**i
        i += 1

    if i == value:
        count += 1

print(count/N)
```

# 8
 Utilizando o método da aceitação/rejeição  gerar amostras para a distribuição f(x)=1.5x2,−1<x<1Plotar a pdf analitica e o histograma normalizado.

```python
import numpy as np
import matplotlib.pyplot as plt

def pdf(x):
    return 1.5 * x**2

# Gerar amostras
N = 100000
accepted = []
while len(accepted) < N:
    x = np.random.uniform(-1, 1)
    y = np.random.uniform(0, 1.5)
    if y < pdf(x):
        accepted.append(x)

# Plotar o histograma normalizado
plt.hist(accepted, bins=50, density=True, alpha=0.5, label='Amostras')

# Plotar a PDF analítica
x = np.linspace(-1, 1, 100000)
y = pdf(x)
plt.plot(x, y, 'r', label='PDF Analítica')

plt.title('Histograma Normalizado e PDF Analítica')
plt.xlabel('Valor')
plt.ylabel('Frequência')
plt.legend()
plt.show()
```

```python
n = 10000
xfunction = np.array([])

for i in range(n):
    x1 = np.random.uniform(-1, 1)
    x2 = np.random.uniform(0, 1.5)

    while (x2 >= pdf(x1)):
        x1 = np.random.uniform(-1, 1)
        x2 = np.random.uniform(0, 1.5)
    xfunction = np.append(xfunction, x1)

X=np.linspace(-1, 1, 100000)
fx = 1.5 * X ** 2
plt.plot(X, fx)
plt.hist(xfunction, bins=100, density=True)
plt.show()
```
